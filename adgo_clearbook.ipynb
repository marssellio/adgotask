```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("testtask.csv")
```

# conversion rate

```python
conv = df.groupby('campaign').agg({'impression':'sum', 'conversion':'sum'})
conv['cr'] = conv.conversion/conv.impression * 100
conv
```

Третья кампания лучше всего конверсируется, но и оплата за нее больше

# cpm 

```python
cpm = df.groupby('campaign').agg({'impression':'sum', 'revenue':'sum'})
cpm['cpm'] = cpm.revenue/conv.impression * 1000
cpm
```

# campaign 1

```python
camp1 = df[df.campaign == 1]
```

# browser
find the highest cr browser

```python
br1 = camp1.groupby('browser').agg({'impression':'sum', 'conversion':'sum'})
br1['cr'] = br1.conversion/br1.impression * 100
br1.sort_values(by='cr', ascending=False)
```

Facebook, Snapchat и Instagram имеют небольшую статистику, поэтому я бы не стал как то серьезно рассматривать их. Возможно, там лучше таргетируется. Любопытно посмотреть подробнее Safari and Chrome. 

```python
camp1[camp1.browser.isin(['Safari','Chrome'])].groupby(['browser','zone']).agg({'impression':'sum', 'conversion':'sum'})
```

У хрома вообще слабая конверсия. Возможно, потому что хромом больше пользуются с компьютера. Сафари хорош на 1ой площадке

```python
camp1[camp1.browser == 'Chrome'].groupby('is_tablet').agg({'impression':'sum'})
```

```python
camp1.groupby('is_tablet').agg({'impression':'sum', 'conversion':'sum'})
```

Если посмотреть статистику по регионам, она в целом ровная

```python
camp1.groupby('region_code').agg({'impression':'sum','conversion':'sum'}).sort_values(by='conversion', ascending=False).head()
```

возможно, есть какая то зависимость от количества показов одному пользователю. 

```python
view1 = camp1.groupby('view_num').agg({'impression':'sum','conversion':'sum'})
view1['cr'] = view1.conversion/view1.impression * 100
view1.sort_values(by='cr', ascending=False).head(7)
```

```python
camp1[camp1.browser.isin(['Safari','Chrome'])].groupby(['browser','is_tablet']).agg({'impression':'sum', 'conversion':'sum'})
```
